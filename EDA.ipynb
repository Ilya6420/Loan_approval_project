```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data_train = pd.read_csv('train.csv')
```

```python
data_train.drop('Loan_ID', axis=1, inplace=True)
```

### First data look analysis:
 1. More Loans are approved Vs Rejected
 2. Count of Male applicants is more than Female
 3. Count of Married applicant is more than Non-married
 4. Count of graduate are more than non-Graduate
 5. Count of self-employed is less than that of Non-Self-employed
 6. Maximum properties are located in Semiurban areas
 7. Credit History is present for many applicants
 8. Count of applicants with number of dependents=0 is maximum.

```python
fig,ax = plt.subplots(2,4,figsize=(16,10))
sns.countplot('Loan_Status',data=data_train,ax=ax[0][0])
sns.countplot('Gender',data=data_train,ax=ax[0][1])
sns.countplot('Married',data=data_train,ax=ax[0][2])
sns.countplot('Education',data=data_train,ax=ax[0][3])
sns.countplot('Self_Employed',data=data_train,ax=ax[1][0])
sns.countplot('Property_Area',data=data_train,ax=ax[1][1])
sns.countplot('Credit_History',data=data_train,ax=ax[1][2])
sns.countplot('Dependents',data=data_train,ax=ax[1][3]);
```

#### The following distribution plot shows that majority of those people who have submitted an application for loan have income between 0 & 10000($).

```python
plt.figure(figsize=(10,7))
sns.distplot(data_train['ApplicantIncome'],color='Black')
plt.xticks(size=12)
plt.yticks(size=12)
plt.xlabel('Applicant_Income', size=12);
```

#### The following distribution plots show that majority of people made a request for a loan for the amount  between 50,000 & 200,000($).

```python
plt.figure(figsize=(10,7))
sns.distplot(data_train['LoanAmount'],color='purple')
plt.xticks(size=12)
plt.yticks(size=12)
plt.xlabel('Loan_Amount', size=12);
```

#### The following illustration indicates that there were more Male candidates for loan application as compared to Females, but the loan status is not much affected by it as there is almost same proportion of rejections in both the cases.

```python
plt.figure(figsize=(10,7))
sns.countplot(x='Gender',data=data_train,hue='Loan_Status',palette='Set2_r')
plt.xticks(size=12)
plt.yticks(size=12)
plt.xlabel('Gender', size=12)
plt.legend(fontsize='large');
```

#### The following illustration outlines that mainly those applicant's application for loan was approved who were not married. So if someone is not married, that person has greater chance for loan to be approved.

```python
plt.figure(figsize=(10,7))
grid=sns.FacetGrid(data=data_train,col='Loan_Status',height=5.0)
grid.map(sns.countplot,'Married',palette='jet_r');
```

#### The following figure indicates that those who had no dependents got their loan approved compared to those who had. So if someone has no dependents, that person has greater chance for loan to be approved.

```python
plt.figure(figsize=(10,7))
sns.countplot('Dependents',data=data_train,hue='Loan_Status',palette='magma')
plt.xticks(size=12)
plt.yticks(size=12)
plt.xlabel('Dependents', size=12);
```

#### The following plot gives an idea about the people's education. It explains that those people who were graduate, had greater chance for their loan to be approved.

```python
plt.figure(figsize=(10,7))
sns.countplot('Education',data=data_train,hue='Loan_Status',palette='inferno_r')
plt.xticks(size=12)
plt.yticks(size=12)
plt.xlabel('Education', size=12);
```

#### The following figure suggests that people who were not self employed were given more preference over those who were self employed. If someone is not self employed, that person has more chance for getting loan approved.

```python
grid=sns.FacetGrid(data=data_train,col='Loan_Status',height=5.0)
grid.map(sns.countplot,'Self_Employed',palette='rocket');
```

#### The following plot shows the relationship between credit history and loan status. It suggests that those who had bad credit history didnt get their loan application approved compared to those who had god credit history. So if anyone has bad credit history, that person might have to face disappointment.

```python
grid=sns.FacetGrid(data=data_train,col='Loan_Status',height=5.0)
grid.map(sns.countplot,'Credit_History',palette='gist_earth');
```

#### This count plot illustrates the property area applicants have. This feature is not showing any relation with loan status as people who had different property area got their applications approved. Through property area it cannot be distinguised that who's application has more chance of being approved.

```python
grid=sns.FacetGrid(data=data_train,col='Loan_Status',height=5.0)
grid.map(sns.countplot,'Property_Area',palette='CMRmap_r');
```

#### The following plot suggests that  the loan was issued for 180 and 360 months

```python
plt.figure(figsize=(15,5))
sns.countplot('Loan_Amount_Term',hue='Loan_Status',linewidth=2.5,edgecolor=".2",data=data_train);
plt.xticks(size=12)
plt.yticks(size=12)
plt.xlabel('Loan_Amount_Term', size=12);
```
